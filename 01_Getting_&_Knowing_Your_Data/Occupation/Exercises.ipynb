### Step 13. What is the most frequent occupation?

```python
# 내가 작성한 답변
users['count'] = 1
users_ = users.groupby('occupation').sum()
users_ = users_.sort_values(['count'], ascending = False)
users_.head(1)
```

### Step 17. What is the mean age of users?

```python
# 내가 적은 답
users__ = users.groupby('user_id').sum()
print(users__.mean()['age'])
```
