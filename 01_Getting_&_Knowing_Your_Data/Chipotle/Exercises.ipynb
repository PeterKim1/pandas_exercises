### Step 9. Which was the most-ordered item? 

```python
# 내가 만든 답
item_col = chipo['item_name'].value_counts()
#print(item_col)
most_ordered = item_col.index[0]
print(most_ordered)
```

```python
c = chipo.groupby('item_name') # item_name 기준으로 그룹화
c = c.sum() # 그룹화 된 것을 sum한 결과
c = c.sort_values(['quantity'], ascending = False)
c.head(1)
```

### Step 11. What was the most ordered item in the choice_description column?

```python
# 내가 만든 답
choice_col = chipo['choice_description'].value_counts()
print(choice_col.index[0])
```

```python
c = chipo.groupby('choice_description').sum()
c = c.sort_values(['quantity'], ascending = False)
c.head(1)
```

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
print(chipo['item_price'].dtype)
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: x.replace('$', '')).astype('float')
```
