### Step 7. View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline

```python
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
discipline.head()
```

### Step 8. Sort the teams by Red Cards, then to Yellow Cards(for discipline)

```python
discipline = discipline.sort_values(by = ['Red Cards', 'Yellow Cards'])
discipline
```

### Step 9. Calculate the mean Yellow Cards given per Team(for discipline)

```python
print(discipline['Yellow Cards'].mean())
```

```python
print(discipline.mean())
print(discipline.mean()['Yellow Cards'])
```

### Step 14. Present only the Shooting Accuracy from England, Italy and Russia

```python
# 내가 적은 답
euro12[euro12['Team'].apply(lambda x: True if x in ['England', 'Italy', 'Russia'] else False)][['Team', 'Shooting Accuracy']]
```
