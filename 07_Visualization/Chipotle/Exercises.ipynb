### Step 5. Create a histogram of the top 5 items bought

```python
chipo.groupby('item_name').sum().loc[:, 'quantity'].sort_values(ascending = False).iloc[:5].plot(kind = 'bar')

plt.title('top 5 items bought')
plt.show()
```
